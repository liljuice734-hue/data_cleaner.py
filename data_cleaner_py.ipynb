```python
import pandas as pd
import re

# 1. GENERATE THE MESSY CSV AUTOMATICALLY
messy_data = """Client Name,Email Address,Phone Number,Account Balance
alice Mwansa ,ALICE@gmail.com,+260971234567,$150.00
alice Mwansa ,ALICE@gmail.com,+260971234567,$150.00
Bob Khumalo,bob@yahoo.com,0965554433,"$2,400.50" """

with open("messy_leads.csv", "w") as f:
    f.write(messy_data)

# 2. THE CLEANING PIPELINE ENGINE
def clean_business_data(input_file, output_file):
    print("🚀 Initiating Data Sanitization Pipeline...")

    df = pd.read_csv(input_file)
    initial_rows = len(df)

    # Drop duplicates
    df.drop_duplicates(inplace=True)

    # Standardize headers
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

    # Clean phone numbers
    if 'phone_number' in df.columns:
        df['phone_number'] = df['phone_number'].astype(str).apply(lambda x: re.sub(r'\D', '', x))

    # Clean emails
    if 'email_address' in df.columns:
        df['email_address'] = df['email_address'].astype(str).str.strip().str.lower()

    # Export clean file
    df.to_csv(output_file, index=False)

    print(f"✅ Cleaned successfully! Output saved to: {output_file}")
    print("\n📊 --- PRISTINE DATA PREVIEW ---")
    print(df)

# Execute everything right now
clean_business_data("messy_leads.csv", "pristine_leads_output.csv")
```
